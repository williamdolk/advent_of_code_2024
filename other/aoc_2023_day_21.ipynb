```python
import numpy as np
import pandas as pd
```

```python
# input_file = 'input_1.txt'
input_file = 'test_input_1.txt'
```

```python
with open(file=input_file, mode='r') as file:
    lines = []

    for line in file:
        lines.append(list(line.strip()))
```

```python
nodes = []
starting_node_found = False
starting_node = None

m = len(lines)

for i in range(m):
    n = len(lines[i])

    for j in range(n):
        current_val = lines[i][j]

        if current_val != '#':
            nodes.append(
                (i, j)
            )

            if not starting_node_found and current_val == 'S':
                starting_node = (i, j)
                starting_node_found = True

nodes.sort()

        
```

```python
A = pd.DataFrame(data=0, columns=nodes, index=nodes)
```

```python
for node in nodes:
    node_right = (node[0], node[1] + 1)
    node_below = (node[0] + 1, node[1])

    if node_right in nodes:
        A.at[node, node_right] = 1
        A.at[node_right, node] = 1

    if node_below in nodes:
        A.at[node, node_below] = 1
        A.at[node_below, node] = 1
```

```python
A
```

```python
number_of_steps = 64
current_nodes = [starting_node]

for step in range(number_of_steps):
    new_current_nodes = set()

    for current_node in current_nodes:
        new_current_nodes.update(set((A.loc[A[current_node] > 0]).index))

    current_nodes = list(new_current_nodes)

len(current_nodes)
```

```python
number_of_steps = 5000
m = len(lines)
n = len(lines[0])

nodes_visited = {starting_node}
current_nodes = [starting_node]
previous_result = 0
current_result = 1

for step in range(1, number_of_steps + 1):
    new_current_nodes = set()

    for node in current_nodes:
        node_left = (node[0], node[1] - 1)
        node_right = (node[0], node[1] + 1)
        node_above = (node[0] - 1, node[1])
        node_below = (node[0] + 1, node[1])

        if (node_left[0] % m, node_left[1] % n) in nodes and node_left not in nodes_visited:
            new_current_nodes.add(node_left)

        if (node_right[0] % m, node_right[1] % n) in nodes and node_right not in nodes_visited:
            new_current_nodes.add(node_right)

        if (node_above[0] % m, node_above[1] % n) in nodes and node_above not in nodes_visited:
            new_current_nodes.add(node_above)

        if (node_below[0] % m, node_below[1] % n) in nodes and node_below not in nodes_visited:
            new_current_nodes.add(node_below)

    current_nodes = list(new_current_nodes)
    nodes_visited.update(new_current_nodes)

    temp_previous_result = current_result
    current_result = previous_result + len(current_nodes)
    previous_result = temp_previous_result
```
